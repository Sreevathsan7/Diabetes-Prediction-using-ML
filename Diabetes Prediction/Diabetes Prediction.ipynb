## Importing the libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Importing the dataset

```python
dataset=pd.read_csv('diabetes.csv')
print(dataset)
```

```python
X=dataset.iloc[:,:-1].values
Y=dataset.iloc[:,-1]
print(X)
print(Y)
```

## Spliting the dataset into training set and testing set

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=0)
```

```python
print(X_train)
```

```python
print(Y_train)
```

```python
print(Y_test)
```

## Feature Scaling on the dataset

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)

## the values in the dataset after scaling are as follows
print(X_train)
print(X_test)
```

# Training the dataset on different Classification models

## Trainig on Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(random_state=0)
lr.fit(X_train,Y_train)
```

```python
ypred1=lr.predict(X_test)
ypred1=ypred1.reshape(-1,1)
y_test=np.array(Y_test)
y_test=y_test.reshape(len(y_test),1)
print(y_test)
```

```python
## Comparing the Predicted value and the actual value from the test set
np.concatenate((y_test,ypred1),1)
```

## Confusion Matrix and accuracy score for the Logistic Regression Model

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm1=confusion_matrix(y_test,ypred1)
acs=accuracy_score(y_test,ypred1)
print(cm1)
print(acs)
```

```python
## The accuracy score we get from Logistic Regression Model is 80% approximately.
```

## Analysing through Clustering

```python
clusdataset=pd.read_csv('Diabetes.csv')

clusdataset=clusdataset[clusdataset['Outcome']==1]

X=clusdataset.iloc[:,[-2,-3]].values
print(X)
```

## Using the elbow method to find optimum number of clusters to be used in K-Means 

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('Elbow method graph for selction of optimum number of clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('WSCC')
plt.show()
```

```python
## from the grpah we can infer that 3 number of clusters would be the number of clusters to use
```

```python
kmeans=KMeans(n_clusters=3,init='k-means++',random_state=42)
y_kmeans=kmeans.fit_predict(X)
print(y_kmeans)
```

## Visualising the results of K-Means Clustering

```python
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 10, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 10, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 10, c = 'green', label = 'Cluster 3')

plt.title('Clusters of Patients Age Group')
plt.xlabel('Age')
plt.ylabel('Diabetes Pedigree Functio')
plt.legend()
plt.show()
```

```python
## By analysing the cluster pattern with Xaxis- Age , Y-axis-Diabetes Predigree Function of all patients who were predicted to
## to be diabetes positive we can infer that the age groups of (20-35),(35-45),(50-70) the age group cluster of (20-30) has more 
## Patients with higher Diabetes Predigree function.
## Diabetes Pedigree Function tells us how likely the patient is to get diabetes heriditarily
## From this we can infer many people in the age group of (20-35) have diabetes very early because of heriditary.
```
